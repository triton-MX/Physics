## Definiciones 

Es necesario anotar algunas definiciones que se consideran útiles para la comprensión del contenido de este cuadernillo: 

- Geodesic = starightest possible path in a curved space/surface. In a curved space, a straight path has **zero tangential acceleration** when traveling along it at a constant speed. [](https://liavas.net/courses/math430/)
- Timelike = this term refers to different separations of events within a given spacetime, more specifically of the interval between two events in spacetime. If the value of the interval between two events is positive, then it is said tha the interval is **time-like**, thus, such events can actually occur at the same location for some observers, and simply be separated in time [Spacelike, Timelike, Lightlike](https://physics.bu.edu/~duffy/ns547_spring10_notes07/Class07_Spacelike.pdf). A more formal definition is, "having overlapping light cones, so that information can pass from one to the other, and one can be envisaged as a cause of the other". [Timelike](https://en.wiktionary.org/wiki/timelike)
- Proper time ($\tau$) = Time measured by a whole collection of observers, each one of them at rest with respect to the others (their own rest frame). Is the time measured by an observer in their own rest frame, i.e the time between two events as measured in a frame where the events stay in the same position. In the case of a particule, proper time would be then the  time given by an imaginary clock strapped to the particle, or a kind of 'internal clock'. [](https://einsteinrelativelyeasy.com/index.php/dictionary/23-proper-time)
- Worldline = The path of a pointlike object in four-dimensional spacetime is a line called the object’s world-line. To **every moment in time corresponds one point of that world-line** giving the position of the object in space at that particular moment.
- Invariant = Any physical quantity is **invariant** when its **value remains unchanged under coordinate or symmetry transformations**. For example, physical quantities represented by scalars, such as temperature, pressure, ...
- Covariant = This term is usually used when the **equations** of **physical systems are unchanged under coordinate transformations**. Physical quantities respresented by **tensors** and appearing in **tensorial equations** are **covariant**. 
- Schwarzschild radius ($r_s$) = or the gravitational radius is a physical parameter in the Schwarzschild solution to Einstein's field equations that corresponds to the radius defining the event horizon of a Schwarzschild black hole. [Referencia](https://en.wikipedia.org/wiki/Schwarzschild_radius)
- Bound orbits = Orbits with a negative total energy are either circular or elliptical in shape. These are known as bound orbits. An orbiting body in a bound orbit cannot escape unless energy is supplied to make the total energy positive. Read the [page](http://www.vikdhillon.staff.shef.ac.uk/teaching/phy105/celsphere/phy105_orbits.html) for more information. 


# Marco Histórico 
## (breve historia sobre agujeros negros)

Actualmente, en pleno siglo XXI, gracias al esfuerzo y labor en conjunto de la comunidad científica, se cuenta con fotografías del agujero negro M87* ubicado en la galaxia Messier 87 a 55 millones de años luz de la tierra, obtenida en 2019, y del agujero negro supermasivo (inserta canción de [Muse](https://www.youtube.com/watch?v=Xsp3_a-PMTw)) Sagitario A* que se encuentra en el centro de nuestra galaxia, la Vía Láctea, adquirida en 2022. Estos objetos, en la última década han sido de gran importancia para el estudio del universo, su origen y su funcionamiento, pero aún se presentan varias dificultades debido a su particular condición gravitacional. Pero, la idea de un agujero negro no es tan reciente como parece, por eso, a continuación se dará a conocer un poco sobre su historia. 

- El concepto de "cuerpos negros" lo suficientemente densos, tanto que la luz no puede escapar de su surperficie, y que por lo tanto se volverían invisibles para el resto del universo, fue pensado por primera vez en **1783 por el filósofo natural inglés John Mitchell**, quién, a pesar de su gran rango de logros científicos, su nombre pasaría regularmente desapercibido. Asimismo, John Michell es conocido como **"el padre de la sismología moderna"**. 

![Michell](https://alchetron.com/cdn/john-michell-3c290600-fd2a-4548-9399-a97a5b64c06-resize-750.jpg)

Esta idea surgió a partir de una consideración hipotética para la obtención de la masa de una estrella (método que posteriormente sería provado erróneo), la cual toma como referencia la teoría de Newton que propone que "la propagación de la luz es causada por el movimiento rectilíneo de partículas de luz, llamadas corpúsculos" ([Referencia](https://scienceready.com.au/pages/particle-and-wave-model-of-light#:~:text=Newton's%20model%20of%20light%20proposes,human%20eye%2C%20obey%20Newtonian%20physics.)). Dicha consideración establece que, la velocidad de los corpúsculos que emergen de la superficie de una estrella, se vería reducida por la atracción gravitatoria de dicha estrella. Entonces, al medir la reducción en la velocidad de la luz de cierta estrella, sería posible calcular la masa de la estrella (es aquí donde, gracias a la teoría de la Relatividad de Einstein formulada en 1905, se sabe que la luz se mueve a través del espacio a una velocidad constante, independientemente de la gravedad local, por lo tanto, este método no sería correcto). De acuerdo con su razonamiento, para cualquier proyectil, éste debería tener una velocidad específica para escapar a la fuerza gravitacional de la estrella (velocidad de escape), así que, ¿qué pasaría si la gravedad de una estrella fuera tan fuerte que la velocidad de escape tuviera que ser mayor a la velocidad de la luz? Entonces, de acuerdo con Michell, la luz tendría que regresar a la superficie.
([Referencia General del texto](https://www.amnh.org/learn-teach/curriculum-collections/cosmic-horizons-book/john-michell-black-holes))

- Independientemente de Michell, Laplace, en el sexto capítulo de "Exposition du Systéme du Monde" en 1976, introdujo una descripción del Sistema Solar, en la cual realiza la siguiente conjetura,

<center>
"La atracción gravitacional de una estrella con un diámetro de 250 veces el del sol comparable con la densidad de la tierra sería de tal magnitud que ni siquiera la luz podría escapar de su superficie. Los cuerpos más grandes del universo serían entonces invisibles debido a su magnitud". 
</center>

<br>

Más adelante, en 1799, ya no solamente planteó lo anterior de forma cualitativa, ahora presentó la demostración matemática de la existencia de dichos "cuerpos invisibles" en forma de un ensayo. 

[Referencia al texto de este cuadro](https://adsabs.harvard.edu/full/2009JAHH...12...90M)

- Realizando un salto temporal hasta el 25 de noviembre de 1915, Albert Einstein publicó las ["Ecuaciones de Campo de la Relatividad General"](https://www.youtube.com/watch?v=M0au5bvEnIA), de la cuales los agujeros negros surgen como una consecuencia predicha por la forma en la que la gravedad distorsiona el espacio-tiempo de tal manera que nada que se acerque a cierta distancia crítica del agujero negro puede escapar, ni siquiera la luz. 

- Finalmente, y con motivo de los artículos estudiados, se hablará sobre el físico y matemático alemán Karl Schwarzschild, quién en 1915 encontró y en 1916 publicó la **solución esféricamente simétrica y estática** a la ecuación de campo en el vacío de Einstein. Dicha solución describe exactamente el comportamiento de la curvatura del espacio-tiempo alrededor de un objeto esférico masivo, sin rotación y sin carga (también conocido como agujero negro de Schwarzschild). De lo más destacado de la solución es que para estrellas muy compactas y de gran densidad, se vuelve muy complicado escapar al campo gravitacional de la estrella, hasta que, eventualmente, cualquier partícula e inclusive la luz no pueden escapar. A este punto se le conoce como el "Horizonte de Eventos".

- Aunque se haya mencionado que el punto anterior sería el final, cabe destacar que, el término **"Agujero Negro"** no fue usado sino hasta **1967** por primera vez por J. Wheeler, supuestamente en una charla en Nueva York. 

# ¿Qué es un agujero negro? 
### Definición y ¿de qué se compone?

Resultaría extraño no hablar sobre el elemento esencial del cual se está partiendo para el desarrollo de este proyecto, así que, ¿qué es un agujero negro? Un **agujero negro** es un área en el espacio donde la gravedad es tan fuerte que inclusive la luz no puede escapar, e inevitablemente se mueve hacia el centro del agujero negro, donde toda la materia se acumula en un solo punto, generando una densidad de masa infinitamente grande, una singularidad. 

Lo anterior es una definición general, y para poder complementarla es necesario introducir otra noción fundamental, esta es el de poder medir masa en unidades de longitud. De acuerdo con la sección *1.1 Matter and geometry, or why you are not attractive (and who is)* del artículo [Black holes in the classical and quantum world](https://arxiv.org/pdf/2306.11139), 

<center>
    "dada una masa M es posible construir una cantidad con dimensiones de longitud que proporcional a esta," 
</center>

$$
L_M = \frac{G}{c^2}M, 
$$

lo que implica que es posible medir masa en metros. Así que, imaginando un objeto con un tamaño específico de $R_o$, que distorsiona de tal manera la geometría a su alrededor que el radio de la curvatura es comparable con el radio $R_o$, lo cual dice, 

$$
Curvatura = \frac{G}{c^2}\text{Densidad de Masa}, 
$$

y las ecuaciones de Eisntein determinan que para que tal objeto exista éste debe ser sumamente masivo y compacto que su tamaño físico $R_o$ es comparable a su masa en unidades de longitud, lo que crea una extensa curvatura en el espacio tiempo, volviéndolo "atractivo". 

![Black Hole](https://science.nasa.gov/wp-content/uploads/2023/09/black_hole_16x9_2.jpg?w=1600)

Ya con una definición y un concepto en mente sobre qué es un agujero negro, se darán a conocer algunos elementos que lo conforman, más específicamente los más importantes para nuestro estudio. 

[Referencia a estos puntos: Black Hole Anatomy](https://science.nasa.gov/universe/black-holes/anatomy/)

- **Horizonte de eventos**: Se puede considerar como la **superficie** del agujero negro. Dentro del horizonte de eventos, la velocidad requerida para escapar del agujero negro excede la velocidad de la luz, por lo tanto, también se puede entender como la región limite en la cual ninguna partícula puede escapar del agujero. Debido a que cualquier partícula, incluyendo fotones, no pueden escapar una vez dentro de esta región, los agujeros negros no reflejan ni emiten luz o algún tipo de radiación (excepto por la posibilidad de un proceso cuántico conocido como la Radiación de Hawking, fenómeno que aun sigue en debate).

- **Disco de Acreción**: La principal fuente de luz de un agujero negro es esta estructura. Los agujeros negros pueden aumentar de tamaño, ya sea consumiendo materia, lo cual se conoce como "acreción" que forma un disco de materia o "disco de acreción" que rodea al agujero, o mediante la fusión con otros agujeros negros. Dicha materia gradualmente se mueve desde la parte exterior del disco hasta su interior, cayendo dentro del horizonte de eventos. A los agujeros negros que han consumido toda la materia que les rodea, y que por tanto no tienen disco de acreción, se les conoce como agujeros negros aislados, y pueden resultar difíciles de estudiar.

Pero, ¿por qué el disco de acreción emite radiación? El primer modelo realista de un disco de acreción fue formulado por Shakura & Sunyaev en 1973. Ellos consideraron que la materia rotando en una órbita circular de Kepler alrededor de cierto objeto compacto **pierde momento angular** debido a la fricción entre las capas adyacentes, causando que caiga en espiral hacia adentro. En el proceso, **energía gravitacional es liberada**, la energía cinética del plasma aumenta y el disco aumenta su temperatura, emitiendo energía térmica y a su vez radiación electromagnética.  ([Referencia a este párrafo](https://www.aanda.org/articles/aa/full_html/2013/03/aa20378-12/aa20378-12.html))


- **Sombra del horizonte de eventos**: Para que se tenga una sombra, es necesario que exista un horizonte de eventos, el cual atrapa la luz y debido a que el espacio-tiempo a su alrededor, mediante una lente gravitacional, distorsiona dicha luz. Estos fenómenos anteriores son los que producen una zona a la que se conoce como "la sombra del horizonte de eventos", la cual es aproximadamente el doble de grande que el propio horizonte de eventos.

- **Singularidad**: La relatividad general predice que un agujero negro contiene en su  centro un punto donde la materia es compactada a una densidad infinita. Es el destino final para cualquier cuerpo o partícula cayendo dentro del horizonte de eventos. Dicha singularidad podría ser una estructura física o meramente matemática, aún se desconoce.

# 3.3 Timelike Geodesics in the Schwarzschild spacetime

## Métrica de Schwarzschild
De acuerdo con las notas estudiadas, el cuerpo de estudio es un agujero negro de Schwarzschild, el cual, como se describió en la parte superior de las notas, es la **solución única esféricamente simétrica y estática** a la ecuación de campo **en el vacío** de Einstein sin constante cosmológica. Esto significa que solamente queda el término $$ R_{\mu\nu}=0$$
donde $R_{\mu\nu} $ es conocido como el tensor de Ricci, por lo tanto, la métrica viene dada por 
haciendo uso de la que ahora es referida como **métrica de Schwarzschild**. Dicha solución identificó el problema esféricamente simétrico de un **punto de masa** con cierto radio, conocido como el **radio de Schwarzschild**$(r_s)$. La masa curva el espacio-tiempo y por tanto afecta a las partículas moviéndose cerca. A continuación se dará una explicación más formal sobre la métrica de Schwarzschild. 

## Continuación del capítulo

Recordando que si, las geodésicas de tipo tiempo (timelike goedesics), se parametrizan por tiempo adecuado (proper time), satisfacen las ecuaciones 

$$0 = \frac{d}{d\tau}\left(\frac{\partial \mathcal{L}(x,\dot{x})}{\partial\dot{x}^\mu} - \frac{\partial \mathcal{L}(x,\dot{x})}{\partial x^\mu} \right),$$

donde, $\mathcal{L}(x,\dot{x}) = \frac{1}{2}g_{\mu v} \: \dot{x}^\mu\dot{x}^v$ , y 

$$
g_{\mu v} \dot{x}^\mu\dot{x}^v = -c^2
$$

Siendo la métrica de Schwarzschild esféricamente simétrica, por lo tanto, se puede especificar que el movimiento se realiza en el plano equatorial, $\theta = \frac
{\pi}{2}$ , $\dot{\theta} = 0$, por ello, dado que la métrica de Schwarzschild en coordenadas esféricas es 

\begin{align}
ds^2 = -c^2d\tau^2 &= -\left(1-\frac{r_s}{r}\right)c^2dt^2 + \frac{dr^2}{1-\frac{r_s}{r}} + r^2(d\theta^2 + \sin^2(\theta)d\phi^2) \\
&= -\left(1-\frac{r_s}{r}\right)c^2\dot{t}^2 + \frac{\dot{r}^2}{1-\frac{r_s}{r}} + r^2(\dot{\theta}^2 + \sin^2(\theta)\dot{\phi}^2).\\ 
\end{align}

Aplicando las condiciones $\theta = \frac{\pi}{2}$ y $\dot{\theta} = 0$, y sustituyendo en la expresión del Lagrangiano, se tiene, 

$$
\mathcal{L}(x,\dot{x}) = \frac{1}{2}\left[-\left(1-\frac{r_s}{r}\right)c^2\dot{t}^2 + \frac{\dot{r}^2}{1 - \frac{r_s}{r}} + r^2\dot{\phi}^2 \right]. 
$$

Utilizando la ecuación de Euler-Lagrange, 

$$ 
\frac{d}{dx}\left(\frac{\partial F}{\partial y'}\right) - \frac{\partial F}{\partial y} = 0
$$
$$
\text{donde y' se interpreta como una de las componentes de la función F}
$$

con $y'$ como $\dot{t}$  
\begin{align}
\frac{d}{d\tau}\left(\frac{\partial \mathcal{L}(x,\dot{x})}{\partial \dot{t}}\right) - \frac{\partial \mathcal{L}(x,\dot{x})}{\partial t} & = 0 \\
-\frac{d}{d\tau} \left[\left(1 - \frac{r_s}{r}\right)c^2\dot{t}\right] - 0 &= 0 \\
\frac{d}{d\tau} \left[\left(1 - \frac{r_s}{r}\right)c^2\dot{t}\right]  &= 0 , 
\end{align}


y $\dot{\phi}$,
\begin{align}
\frac{d}{d\tau}\left(\frac{\partial \mathcal{L}(x,\dot{x})}{\partial \dot{\phi}}\right) - \frac{\partial \mathcal{L}(x,\dot{x})}{\partial \phi} & = 0 \\
\frac{d}{d\tau} \left(r^2\dot{\phi}\right) - 0 &= 0 \\
\frac{d}{d\tau} \left(r^2\dot{\phi}\right) &= 0 . 
\end{align}

se obtienen las siguientes **constantes de movimiento**, la **energía (E)** y el **momento angular (L)**, 

\begin{align}
\frac{d}{d\tau} \left[\left(1 - \frac{r_s}{r}\right)c^2\dot{t}\right]  &= 0 , \;\;\; E = \left(1-\frac{r_s}{r}\right)c^2\dot{t} = \text{constante} \\
\frac{d}{d\tau} \left(r^2\dot{\phi}\right) &= 0, \;\;\; L = r^2\dot{\phi} = \text{constante} , 
\end{align}

y, la definición de tiempo propio requiere que se cumpla la siguiente expresión 
$$
-\left(1-\frac{r_s}{r}\right)c^2\dot{t}^2 + \frac{\dot{r}^2}{1 - \frac{r_s}{r}} + r^2\dot{\phi}^2 = -c^2. 
$$
Es através de las tres ecuaciones anteriores que se puede obtener que la componente $\dot{r}$, que aún es desconocida. Despejando $\dot{t}$  y $\dot{\phi}$ a partir de las expresiones de las constantes se tiene que, 

\begin{align}
\dot{t} &= \frac{E}{c^2\left(1-\frac{r_s}{r}\right)}, \\
\dot{\phi} &= \frac{L}{r^2}, 
\end{align}

y, utilizando la expresión definida para el tiempo propio, se despeja $\dot{r}$, 
\begin{align}
-\left(1-\frac{r_s}{r}\right)c^2\left[\frac{E}{c^2\left(1-\frac{r_s}{r}\right)}\right]^2 + \frac{\dot{r}^2}{1 - \frac{r_s}{r}} + r^2\left(\frac{L^2}{r^4}\right) &= -c^2 \\
-\frac{E^2}{\left(1 - \frac{r_s}{r}\right)c^2} + \frac{\dot{r}^2}{1 - \frac{r_s}{r}} + \frac{L^2}{r^2} &= -c^2, \\ 
\end{align}
donde, 
\begin{align}
\dot{r}^2 &= \left(1 - \frac{r_s}{r}\right)\left[\frac{E^2}{\left(1 - \frac{r_s}{r}\right)c^2} - \frac{L^2}{r^2} - c^2\right] \\ 
& = \frac{E^2}{c^2} - \left(1 - \frac{r_s}{r}\right)\left(\frac{L^2}{r^2} + c^2 \right)
\end{align}

De la ecuación anterior se puede definir al **potencial efectivo ($V_{eff}$)** como, 
<a id="eq_Potencial"></a>
$$
V_{eff}^2 = \left(1 - \frac{r_s}{r}\right)\left(\frac{L^2}{r^2} + c^2 \right).
$$
Entonces, sean la primera y segunda derivada del potencial efectivo con respecto al tiempo propio ($\tau$), 
$$
V'_{eff} = \frac{3r_sL^2}{r^2} - \frac{2L^2}{r} - r_sc^2, 
$$
y 
$$
V''_{eff} = 2L^2 - \frac{6r_sL^2}{r}. 
$$
Debido a que se introdujo al potencial efectivo ($V_{eff}$) de tal manera que se cumple y se mantiene una expresión análoga a la "conservación de le energía", 
$$
\frac{1}{2}\left(\frac{dr}{d\phi}\right)^2 + V_{eff} = 0. 
$$
Y, para tener una órbita con constantes de movimiento E y L, éstas deben estar confinadas a la región donde $V_{E,L}(r)\leq 0$; los puntos límite, donde $V_{E,L}(r) = 0$, son puntos de retorno de la órbita donde $\frac{dr}{d\phi} = 0 $. Entonces, a través del protencial efectivo es posible obtener las expresiones para los valores de E y L, donde **órbitas ligadas (bound orbits) y órbitas circulas estables o inestables** existen. Por lo tanto, se establecen las siguientes condiciones: 
\begin{align}
V_{eff}' &= \frac{3r_sL^2}{r^2} - \frac{2L^2}{r} - r_sc^2 = 0 \;\;\;\; (i)\\
V_{eff}'' &= 2L^2 - \frac{6r_sL^2}{r}  = 0 \;\;\;\;\;\;\; (ii)\\
\dot{r}^2 &= \frac{E^2}{c^2} - \left(1 - \frac{r_s}{r}\right)\left(\frac{L^2}{r^2} + c^2 \right) \\
& = \frac{E^2}{c^2} - V_{eff}^2 = 0 \;\;\;\;\;\;\;\;\;\;  (iii)
\end{align}

Es posible observar que, a partir de la expresión $(ii)$ se puede obtener **La Órbita Circular Estable más Interna** o **ISCO** por sus siglas en inglés, recordando que la segunda derivada igualada a cero nos permite identificar si un punto crítico es un mínimo o un máximo. La condición de estabilidad se cumple si, 
\begin{align}
V_{eff}'' = 2L^2 - \frac{6r_sL^2}{r} &= 0 \\
L^2(2 - \frac{6r_s}{r}) &= 0 \\
r &= 3r_s, 
\end{align}
que se conoce como el ISCO. Una partícula masiva no puede tener una órbita estable alrededor de un agujero negro de Schwarzschild si se encuentra a un radio menor que $3r_s$ (punto crítico mínimo). Dicho punto crítico también se puede observar cuando se grafica la función. 

```python
from scipy.optimize import minimize_scalar
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

def potencial(r,L): #r = r gorrito
    veff = (1-(1/r))*((L**2/r**2)+1)
    return veff

r = np.linspace(1,100,100)
valor = minimize_scalar(lambda r: potencial(r, 6**(1/2)), bounds=(1, 100), method='bounded')

print("Minimum point (r):", valor.x) #finds the x value at which it is minimum
print("Minimum value (potencial):", valor.fun) # evaluates the function at said x - minimum value (yes sir)

plt.ylim(0.8, 1.3)
plt.plot(r, potencial(r, 6**(1/2)), label='Potencial', color='red')
plt.plot(valor.x, valor.fun, 'bo', label='Punto Mínimo')
plt.grid()
plt.legend()
plt.xlabel('r')
plt.ylabel('Potencial')
plt.show()
```

A partir de la igualdad $(i)$ es posible obtener la expresión para el momento angular ($L$). 
\begin{align}
  -\frac{2L^2}{r} + \frac{3r_sL^2}{r^2} + r_sc^2 &= 0, \\ 
  L^2\left(\frac{3r_s}{r^2} - \frac{2}{r}\right) + r_sc^2 &= 0, 
\end{align}
despejando $L^2$, se tiene, 
<a id="eq_Angular"></a>
$$
L^2 = \frac{-r_sc^2}{\left(\frac{3r_s}{r^2} - \frac{2}{r}\right)} = \frac{r^2r_sc^2}{2r - 3r_s}. 
$$
Después, de $(iii)$ se despeja $E$. Se sustituye el valor para el momento angular al cuadrado y se recuerda la expresión para el [Potencial efectivo al cuadrado](#eq_Potencial), 

\begin{align}
  V_{eff}^2 &= \frac{E^2}{c^2} \\ 
  L^2\left(\frac{1}{r^2} - \frac{r_s}{r^3}\right) + c^2\left(1 - \frac{r_s}{r}\right) &=   \frac{E^2}{c^2} \\
  \left(\frac{r^2r_sc^2}{2r - 3r_s}\right)\left(\frac{1}{r^2} - \frac{r_s}{r^3}\right) + c^2\left(1 - \frac{r_s}{r}\right) &=   \frac{E^2}{c^2} \\
  \frac{r_sc^2(r-r_s) + (2r - 3r_s)c^2(r-r_s)}{r(2r - 3r_s)} &= \frac{E^2}{c^2} \\
  c^4(r-r_s)\left[\frac{r_s + (2r - 3r_s)}{r(2r - 3r_s)}\right] &= E^2.\\
\end{align}
Entonces, la expresión final para la Energía está dada por: 
<a id="eq_Energía"></a>
$$
 E^2 = \frac{2c^4(r-r_s)^2}{r(2r-3r_s)}
$$

Una vez obtenidos los valores para las constantes anteriores, se realiza la sustitución con la finalidad de simplificar la expresión para poder graficarla, 
$$
\tilde{r}_s = \frac{r}{r_s},
$$
de ahí se despeja $r$, 
$$
r_s = \frac{r}{\tilde{r}}.
$$
Mediante este cambio de variable, se puede expresar el [Momento Angular al cuadrado](#eq_Angular), y la [Energía al cuadrado](#eq_Energía) de la forma siguiente: 
\begin{align}
  L^2 &= \frac{\tilde{r}^2}{2\tilde{r} - 3}, \\
  E^2 &= \frac{2(\tilde{r}-1)^2}{r(2\tilde{r}-1)}.
\end{align}

# Disk Model and the Power Spectrum

A continuación se darán a conocer algunas propiedades del disco, como $D(r)$, la temperatura $T(\nu)$ y la Luminosidad $L(\nu)$. 

Primero, se define a $D(r)$ como la razón por unidad de área en la cual la energía cinética es disipada en calor por viscosidad o también como el trabajo por unidad de área generado por el disco de acreción rotando alrededor de un objeto. Así, 
<a id="Dr"></a>
$$
D(r) = \frac{\dot{M}}{4\pi r}\frac{\alpha}{a}\left(-\frac{d\Omega}{dr}\right)\frac{1}{(E - \Omega L)^2}\int_{r_i}^{r_f} (E-\Omega L)\frac{dL}{dr}dr ,
$$
donde, $\dot{M}$ es la tasa de masa de acreción y $r_i$ es el borde interior del disco de acreción. Para agujeros negros, este radio $r_i$ se asume que está en el *ISCO* ($r=6M$) del propio agujero. 

Haciendo uso de las expresiones al cuadrado de la [energía](#eq_Energía) y del momento [angular](#eq_Angular), además de algunas manipulaciones algebraicas, se puede obtener una expresión para $D(r)$ posible de graficar. 

Dentro de la información contenida en el artículo [Spherical boson stars as black hole mimickers](https://arxiv.org/abs/1009.1250) se explica que, en las geodésicas para un agujero negro de Schwarzschild se tienen los valores, 

$$
\alpha^2 = a^{-2} = 1 - \frac{2M}{r},
$$
donde $\alpha$ es la función de lapso (las ecuaciones de Einstein permiten cierta arbitrariedad cuando uno elige foliar el espacio-tiempo en una hipersuperficie espacial tridimensional con una dirección ortogonal del tiempo. Dos funciones, el lapso y el desplazamiento, indican cómo relacionar coordenadas entre dos segmentos: el lapso $\alpha$
**mide el tiempo propio/adecuado (proper time)**) y $a$ es la función métrica radial. Así que, retomando la ecuación para [D(r)](Dr) se tiene que, 

$$
\frac{\alpha}{a} = 1 - \frac{2M}{r}. 
$$

También, la velocidad angular ($\Omega$) de una partícula de prueba es
$$
\Omega = \frac{\dot{\phi}}{\dot{t}}, \;\;\; \text{donde} \;\; \dot{t} = \frac{E}{\left(1-\frac{r_s}{r}\right)c^2} \;\; y  \;\; \dot{\phi} = \frac{L}{r^2}. 
$$
Así, y utilizando $\ell$ en lugar de $L$ para evitar confundirla con la luminosidad, 
$$
\Omega = \frac{\ell\left(1-\frac{r_s}{r}\right)}{Er^2}.
$$
Recordando la geometrización de las unidades $c=1$, que el parámetro $r_s = 2M$, y sustituyendo las expresiones al cuadrado de la [energía](#eq_Energía) y del momento [angular](#eq_Angular), se obtiene,
$$
\Omega = \frac{\sqrt{Mr}}{r^2}. 
$$
Además, 
$$
-\frac{d\Omega}{dr} = \frac{3\sqrt{Mr}}{2r^3}. 
$$
Otro elemento a calcular es también $(E - \Omega L)^2$, así que, partiendo de 
\begin{align}
E - \Omega L &= \frac{\sqrt{2}(r-r_s)}{\sqrt{r(2r - 3r_s)}} - \frac{\sqrt{2r_sr}}{2r^2}\left(\frac{r\sqrt{r_s}}{\sqrt{2r-3r_s}}\right) \\
&= \frac{\sqrt{2(2r-3r_s)}}{2\sqrt(r)}. 
\end{align}
Sustituyendo $r_s = 2M$, se tiene 
$$
E - \Omega L = \frac{\sqrt{2(2r-6M)}}{2\sqrt{r}} = \frac{(r-3M)}{\sqrt{r(r-3M)}}, 
$$
y elevando la expresión al cuadrado se obtiene 
$$
(E - \Omega L)^2 = \frac{2(2r-6M)}{4r} = \frac{(r-3M)}{r}. 
$$
Después, se va a obtener  
$$
\frac{dL}{dr} = \frac{d}{dr}\left(\frac{r\sqrt{r_s}}{\sqrt{2r-3r_s}}\right) = \frac{\sqrt{r_s}(r-3r_s)}{(2r-3r_s)^{3/2}}, 
$$
sustituyendo $r_s = 2M$ se tiene, 
$$
\frac{dL}{dr} = \frac{\sqrt{2M}(r-6M)}{(2r-6M)^{3/2}}
$$
Ahora
$$
(E-\Omega L)\frac{dL}{dr} = \frac{(r-3M)}{\sqrt{r(r-3M)}}\left(\frac{\sqrt{2M}(r-6M)}{(2r-6M)^{3/2}}\right) = \frac{\sqrt{2M}(r-6M)}{2\sqrt{r}(r-3M)}. 
$$
Una vez obtenidos todos estos segmentos, se puede escribir la expresión completa. Así, 
\begin{align}
D(r) &= \frac{\dot{M}}{4\pi r}\left(1 - \frac{2M}{r}\right)\left(\frac{3\sqrt{Mr}}{2r^3}\right)\left(1-\frac{3M}{r}\right)^{-1}\int_{r_i}^{r_f} \frac{\sqrt{2M}(r-6M)}{2\sqrt{r}(r-3M)} dr, \\
&= -\frac{3\dot{M}M}{16\pi r^{7/2}}\left(1 - \frac{2M}{r}\right)\left(1-\frac{3M}{r}\right)^{-1}\int_{r_i}^{r_f} \frac{6M - r}{\sqrt{r}(r-3M)} dr. 
\end{align}

La expresión anterior se va a graficar utilizando el método de Simpson para Integrales Numéricas, tomando $r_i = 6$ Y $r_f$ cualquier valor mayor a 6. 

```python
integral=[]
X1=[]
def Simpson(a,b,n,f):
    h = (b-a)/n
    XI0 = f(a) + f(b)
    XI1 = 0
    XI2 = 0
    for i in range(1,n): #Es hasta n, porque no toma el último elemento, entonces por eso n -> n-1, y n-1 -> n-2
        X = a + i*h
        if i%2 == 0:
           XI2 = XI2 + f(X)
        else:
           XI1 = XI1 + f(X)
        XI= h*(XI0 + 2*XI2 + 4*XI1)/3 #eliminar esta linea y la de abajo para obtener un único valor
        integral.append(XI)
        X1.append(X)
    return integral,X
```

Para el caso de la **Temperatura**, se dice que la materia que se encuentra en el disco de acreción está en equilibrio termodinámico. Esto implica que el modelo de radiación de cuerpo negro es ideal para describir la radiación emitida en la superficie del disco.  

Dicha temperatura está dada por la expresión: 
$$
T(r) = \left(\frac{D(r)}{\sigma}\right)^{1/4}, 
$$
donde $\sigma = 5.67\times10^{-15} \text{erg s}^{-1} cm^{-2} K^{-4}$ es la constante Stefan-Boltzmann, y la cual se obtiene a partir de la Ley de Stefan-Boltzmann, la cual establece una relación empírica entre la potencia radiada por un cuerpo negro por unidad de área y la temperatura, y se formula: 

$$
D(r) = \sigma T^4
$$ 

Y debido a que no es sencillo encontrar su valor de forma analítica, en la siguiente figura se muestra una gráfica de los datos de temperatura que se obtuvieron. 

## Gráfica de la temperatura

```python
def T(r,integral): 
    Mp = 0.472e+19
    sigma=5.67e-5
    D =(3*Mp/(sigma*16*math.pi*r**3.5))*(r-2/r-3)
    y=(D*integral)**0.25
    return y
valores=pd.Series(np.linspace(6,99,999))
y=pd.Series(T(valores,df['resultado']))
T_r=pd.DataFrame({'r':valores,'T(r)':y})
print(T_r)
plt.plot(T_r['r'],T_r['T(r)'],color='purple')
plt.grid()
plt.title('Temperatura')
plt.xlabel('r')
plt.ylabel('T(r)')
```

Una vez obtenidos los valores de la Temperatura que dependen de $r$, se hablará sobre la **Luminosidad**. La luminosidad es la luz que emite el disco de acreción, es por ello que ésta depende de la frecuencia ($\nu$), pues es la frecuencia de los fotones emitidos. Esta propiedad va de la mano con la temperatura del disco ya que se toma como base el modelo de **radiación de cuerpo negro**, el cual describe la relación entre la **temperatura de un cuerpo** y la **longitud de onda** de la radiación electromagnética que éste emite. Un parámetro importante a considerar es la inclinación del disco, puesto que a mayor inclinación, se cambia el espectro de potencia a luminosidades más altas y a menor inclinación a luminosidades más bajas. 

Un **cuerpo negro** es un objeto idealizado que absorbe todo tipo de radiación electromagnética con la que entra en contacto. Después, emite radiación térmica mediante un espectro continuo de acuerdo con su temperatura.   Los agujeros negros son considerados cuerpos negros **casi perfectos**, debido a que, como ya se ha repetido varias veces en este cuadernillo, abosrben toda la radiación con la que entran en contacto. Se ha sugerido que los agujeros negros emiten radiación de cuerpo negro conocida como **Radiación de Hawking**, sin embargo, esta radiación no ha sido observada o detectada por el momento. 

La luminosidad de un cuerpo negro está dada por la **ley de Stefan-Boltzman**
$$
L = 4\pi R^2\sigma T^4 
$$
donde $\sigma$ es la constante de Boltzmann, T es la temperatura del cuerpo negro (que en este caso el la temperatura del agujero negor) y R es la distancia de la luminosidad (o, para el caso del agujero negor, el radio del horizonte de eventos). Por tanto, y utilizando la dependencia de T a la coordenada radial de la Ley de Boltzmann, se puede obtener la luminosidad usando la expresión para la distribución espectral planteada por Planck, 
$$
R(\mu) = \frac{2\pi\nu^2}{c^2}\frac{h\nu}{e^{\frac{h\nu}{KT}}-1}. 
$$
Así, considerando que el disco de acreción emite como si fuera un cuerpo negro, la luminosidad se puede calcular a través de, 
$$
L = 4\pi R^2\sigma T^4  = 4\pi d^2F(\nu) = \frac{16\pi h}{c^2}\cos(\vartheta)\nu^3\int_{r_i}^{r_f} \frac{rdr}{e^{\frac{h\nu}{KT}}-1}. 
$$
De la expresión anterior se tiene que $d$ es la distancia a la fuente, $r_i$ y $r_f$ indican la ubicación de los bordes interno y externo del disco de acreción, $h = 6.6256 \times 10^{-27}$ $\text{erg s}$ es la **constante de Planck**, $k$ es la ya antes mencionada constante de Boltzmann y $\vartheta$ es la inclinación del disco. 

Y nuevamente, como es una expresión complicada de calcular de forma analítica, se hará uso de métodos numéricos para obtener resultados. La siguiente figura muestra una gráfica de luminosidad que depende de la frecuencia emitida de los fotones.

## Gráfica de luminosidad

```python
temperaturas=T_r[('T(r)')].to_numpy()
integral_p=[]
r1=[]
def Simpson(a,b,n,f,v):
    h = (b-a)/n
    XI0 = f(a,v) + f(b,v)
    XI1 = 0
    XI2 = 0
    for i in range(1,n): #Es hasta n, porque no toma el último elemento, entonces por eso n -> n-1, y n-1 -> n-2
        X = a + i*h
        if i%2 == 0:
           XI2 = XI2 + f(X,v)
        else:
           XI1 = XI1 + f(X,v)
        XI= h*(XI0 + 2*XI2 + 4*XI1)/3 #eliminar esta linea y la de abajo para obtener un único valor
    return XI
valores_p=np.linspace(1e5,1e17,998)
for i in range(len(valores_p)):
    def lum(r,v):
     h=6.6256e-27
     k=1.3805e-16
     t=((h*v)/(k*temperaturas[i]))  
     e=np.exp(t)
     f=r/(e-1)
     return f
    integral_p.append((Simpson(6,100,1000,lum,valores_p[i])))
#for j in range(len(integral_p)):
 #print(integral_p[j])
```

```python
resultados_f=[]
for j in range(len(valores_p)):
 def L(v):
  h=6.6256e-27
  c=3
  l=(16*np.pi*h)/(c**2)
  i=l*np.cos((np.pi)/4)*(v**3)*integral_p[j]
  return i
 resultados_f.append(L(valores_p[j]))

#print(resultados_f)
```
